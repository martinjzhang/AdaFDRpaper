```python
variant_dict = {}
with open(os.path.join(utils_path, 'GTEx_Analysis_2016-01-15_v7_WholeGenomeSeq_635Ind_PASS_AB02_GQ20_HETX_MISS15_PLINKQC.lookup_table.txt')) as f:
    ln = 0
    for line in f:
        if ln > 0:
            ls = line.strip().split()
            if ls[-1] != '.':
                variant_dict[ls[2]] = ls[-1]
        ln += 1
        if ln % 100000 == 0:
            print(ln)
```

```python
len(variant_dict)
```

```python
feature_dict = {}
with open(os.path.join(utils_path, 'common_all_20180423.vcf')) as f:
    ln = 0
    for line in f:
        
        ls = line.strip().split('\t')
        if len(ls) > 2:
            if ls[2][:2] == 'rs':
                #print(ls)
                variant_id = ls[2]
                info = ls[-1].split(';')
                match = [s for s in info if 'CAF=' in s]
                if len(match) > 0:
                    info = match[0][4:]
                info = info.split(',')
                info = [item for item in info if item != '.']
                info = sorted(map(float, info), reverse = True)
                if len(info) > 1:
                    maf = info[1]
                    feature_dict[variant_id] = maf
        ln += 1
        if ln % 100000 == 0:
            print(ln)
```

```python
num = 0
for k,v in feature_dict.items():
    print(k,v)
    num += 1
    if num > 100:
        break
```

```python
num = 0
with open(os.path.join(utils_path, 'snp_feat_maf.txt'), 'w') as f:
    for k,v in variant_dict.items():
        #print(k,v)
        f.write("{},{},{}\n".format(k,v,feature_dict.get(v, 'NA')))
        num += 1
        if num % 100000 == 0:
            print(num)
```
